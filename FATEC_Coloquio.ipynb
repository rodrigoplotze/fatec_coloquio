```
import cv2 as cv                  
import matplotlib.pyplot as plt   

# Carregar a imagem
img1 = cv.imread("fachada.jpg")
img1 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)

# Converter a imagem para Tons de Cinza
img2 = cv.cvtColor(img1,cv.COLOR_RGB2GRAY)

# Converter a imagem para Preto e Branco
_, img3 = cv.threshold(img2, 127, 255, cv.THRESH_OTSU)


# Exibir a imagem
plt.figure(figsize=(10,10))
plt.subplot(311), plt.imshow(img1)
plt.subplot(312), plt.imshow(img2, cmap='gray')
plt.subplot(313), plt.imshow(img2, cmap='binary')
```
